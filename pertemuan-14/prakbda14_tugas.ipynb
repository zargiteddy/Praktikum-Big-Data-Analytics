```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("iris.csv")
print (data.head(5))
```

```python
ax = sns.scatterplot(x='PetalWidthCm', y='PetalLengthCm', data=data)
```

```python
ax = sns.lineplot(x='PetalWidthCm', y='PetalLengthCm', data=data)
```

```python
az = sns.relplot(x='PetalWidthCm', y='PetalLengthCm', data=data)
```

```python
with sns.axes_style('white'):
    sns.jointplot("PetalWidthCm", "PetalLengthCm", data, kind='kde');
```

```python
sns.barplot(x='PetalWidthCm', y='PetalLengthCm', data=data)
```

```python
sns.factorplot(x='PetalWidthCm', y='PetalLengthCm', data=data)
```
