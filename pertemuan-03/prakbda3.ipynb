## 1. Menggunakan fungsi Membuat Ploting data pada sumbu ordinat

### a) Membuat garis koordinat untuk ploting data menggunakan perintah dibawah ini:

```python
import matplotlib.pyplot as plt
plt.axis([0,10,0,25])
```

Sebelum kita menulis kode di atas, pertama-tama kita harus meng-import Matplotlib. Matplotlib adalah library yang paling sering digunakan oleh para data scientist karena dapat digunakan untuk memvisualisasikan data (misalnya dalam bentuk grafis) dalam bentuk 2D atau 3D. Pada kode di atas, kita integrasikan matplotlib dengan pyplot dengan menuliskan matplotlib.pyplot. Baris kode tersebut berfungsi untuk meng-import sebuah koleksi berisi fungsi-fungsi yang membuat matplotlib bekerja sama seperti MATLAB. 

Setiap fungsi pyplot membuat perubahan pada setiap figur: misalnya membuat figur, membuat area plotting di dalam sebuah figur, menambahkan garis pada area plotting, mendekorasi plot dengan label, dan lain-lain. Baris kode `matplotlib.pyplot` tersebut kita dekalrasikan dengan nama plt. Setelah itu, kita integrasikan plt dengan method axis dengan menuliskan `plt.axis([0,10,0,25])`. Kegunaan dari method axis tersebut adalah untuk mengatur limit x dan y, dengan melemparkan list yang memiliki struktur spesifik seperti [xmin, xmax, ymin, ymax].

### b) Menampilkan data ordinat untuk data {(2,4), (2.5,11), (6,14), (7,17), (9,24)} gunakan coding dibawah ini**

```python
plt.plot([2,2.5,6,7,9],[4,11,14,17,24],'gs')
plt.show()
```

Pada kode di atas kita menampilkan data dengan koordinat untuk data {(2,4), (2.5,11), (6,14), (7,17), (9,24)} dengan fungsi `plt`. dengan kode `plt.plot([2,2.5,6,7,9],[4,11,14,17,24],'gs')`. Plt.plot adalah fungsi untuk menampilkan/menggambar titik/point(markers) pada sebuah diagram. `gs` pada kode di atas berarti `GridSpec` yang digunakan untuk menentukan geomerti pada sebuah grid untuk menempatkan sebuah subplot. Lalu untuk menampilkan grafik, bisa kita gunakan baris kode `plt.show()`.

### d) Membuat ploting data dengan memberi nama variabel sumbu x dan sumbu y, seperti dibawah ini

Untuk melakukan itu, kita menggunakan fungsi `xlabel()` dan `ylabel()`. Kedua fungsi tersebut merupakan method di dalam module pyplot dalam library matplotlib yang berguna untuk mengatur label unuk x-axis dan y-axis. Kita tuliskan baris kode `plt.xlabel('Actual birth weight')` dan `plt.ylabel('Estimated birth weight')`. Terakhir, kita tampilkan hasil dengan baris kode `plt.show()`.

### b) Scatter Plot Data

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

insurance = 'insurance.csv'
insurance_data = pd.read_csv(insurance)
insurance_data.head()
```

Link dataset: https://github.com/stedy/Machine-Learning-with-R-datasets/blob/master/insurance.csv

**Fungsi untuk membuat scatter plot**

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

Pada kode di atas kita menggunakan seaborn yang tadi sudah kita deklarasikan dengan nama sns. Dengan sns, kita gunakan scatter plot yang gunanya adalah menampilkan sebuah diagram yang dimana setiap value di dataset direpresentasikan dengan tanda titik berwarna biru. Pada diagram tersebut ada dua data yang kita ambil dan kita tampilkan dari dataset insurance, yaitu data ‘bmi’ pada sisi x dan data ‘charges’ pada sisi y. BMI sendiri adalah sebuah kependekan dari _Body Mass Index_ yang berarti sebuah perkiraan lemak pada tubuh berdasarkan tinggi dan berat badan pada laki-laki dan perempuan dewasa. Sementara ‘charges’ adalah biaya asuransi setiap person. Kode yang kita gunakan untuk menampilkan diagram tersebut adalah dengan menuliskan `sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])`.

**Untuk membuat plotting regresi dengan variabel x = bmi dan y = charges**

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

Pada kode di atas kita menggunakan method `regplot()` yang berguna untuk mem-plot data dan model regresi linear. Data yang kita tampilkan dengan regplot adalah ‘bmi’ dan ‘charges’. Untuk melakukan itu, kita bisa menuliskan baris kode `sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])`.

**Menampilkan plot data kategori dari salah satu variabel smoker**

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'],
            hue = insurance_data['smoker'])
```

Pada kode ini kita mengambil dan menampilkan satu data baru dari dataset insurance yaitu data boolean (true/false) yaitu data smoker. Jika value smoker adalah ‘yes’, maka dot berwarna biru. Jika value smoker adalah ‘no’, maka dot berwarna orange. Kita bisa melakukan ini dengan kode `sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue = insurance_data['smoker'])`.

**Menampilkan prediksi data tentang smoker melalui garis regresi**

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data = insurance_data)
```

Pada tahap ini, kita menggunakan garis regresi dengan fungsi lmplot(). Fungsi tersebut adalah sebuah fungsi seaborn berupa scatterplot dua dimensi yang menampilkan regresi linear overlaid yang beruguna untuk membandingkan variabel numerik. Untuk melakukan itu bisa kita tuliskan `sns.lmplot(x='bmi', y='charges', hue='smoker', data = insurance_data)`.
