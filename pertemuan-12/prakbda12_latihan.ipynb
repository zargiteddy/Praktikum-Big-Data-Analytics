### 1. Kita import library untuk memproses data dan plotting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Untuk menggunakan klasifikasi Support Vector Machine, kita harus mengimport modul “svm” dari SciKit-Learn.

```python
import sklearn.svm as svm
```

### 3. Kita load dataset iris lalu masukkan ke dalam DataFrame menggunakan fungsi pandas

```python
df = pd.read_csv('iris.csv')
df.head()
```

### 4. Kita buat fungsi bernama name_to_number(name) yang berguna untuk mengubah nama bunga (string) menjadi angka supaya bisa diproses oleh model machine learning yang nanti kita buat. Bunga “Iris-setosa” diubah menjadi angka 0, sementara bunga “Iris-versicolor” diubah menjadi angka 1.

```python
def name_to_number(name):
    if name == 'Iris-setosa':
        return 0
    if name == 'Iris-versicolor':
        return 1
    return 2
```

### 5. Karena kita ingin membuat model 2D, kita ambil dua kolom saja yaitu PetalLengthCm dan PetalWidthCm. Kemudian sekaligus kita lihat shape dari X dan y.

```python
X = np.array(df[['PetalLengthCm', 'PetalWidthCm']])
y = np.array([name_to_number(name) for name in df['Species']])
```

```python
X.shape, y.shape
```

### 6. Untuk melakukan plot pada decision boundary pada linear model, kita definisikan fungsi def make_meshgrid(x, y, h=0.005). Fungsi ini mengambil dua vektor yaitu x dan y sebagai input. Parameter h adalah stepsize dalam proses plotting ini. Fungsi ini me-return pasangan (x, y) di mana x dan y adalah matriks dengan shape yang sama.

```python
def make_meshgrid(x, y, h=0.005):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    return np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

### 7. Fungsi def plot_contour(M, X, Y) digunakan untuk mem-plot decision boundaries untuk classifier M. X dan Y adalah meshgrids untuk axis x dan y.

```python
def plot_contour(M, X, Y):
    Z = M.predict(np.c_[X.ravel(), Y.ravel()])
    Z = Z.reshape(X.shape)
    plt.contour(X, Y, Z)
```

### 8. Dengan model M dan matriks 2D X, fungsi ini mem-plot data dari X dan decision boundary.

```python
def plot_data_and_boundary(X, M, title):
    X0, X1 = X[:, 0], X[:, 1]
    XX, YY = make_meshgrid(X0, X1)
    plt.figure(figsize=(15, 10))
    sns.set(style='darkgrid')
    plot_contour(M, XX, YY)
    plt.scatter(X0, X1, c=y, edgecolors='k')
    plt.xlim(XX.min(), XX.max())
    plt.ylim(YY.min(), YY.max())
    plt.xlabel('Petal length')
    plt.ylabel('Petal width')
    plt.xticks()
    plt.yticks()
    plt.title(title)
```

### 9. Pertama, kita mulai dengan model linier. Regularisasi parameter di-set ke 100000, yang berarti tidak ada regularisasi.

```python
M = svm.SVC(kernel='linear', C=100000)
M.fit(X, y)
M.score(X, y)
```

```python
plot_data_and_boundary(X, M, 'Support Vector Machine dengan Linear Kernel')
```

Kelas SVC tersebut menggunaakan classifier One-vsOne yang membuat tiga support vector machines.
- SVM pertama memisahkan setosa dari versicolor
- SVM kedua memisahkan setosa dari virginica
- SVM ketiga memisahkan virginica dari versicolor

### 10. Selanjutnya kita buat model Gaussian kernel

```python
M = svm.SVC(kernel='rbf', gamma=1.5, C=10000)
M.fit(X, y)
M.score(X, y)
```

```python
plot_data_and_boundary(X, M, 'Support Vector Machine dengan Gaussian Kernel')
```

### 11. Buat model Polynomial Kernel of Degree 3

```python
M = svm.SVC(kernel='poly', degree=3, gamma='auto', C=10000)
M.fit(X, y)
M.score(X, y)
```

```python
plot_data_and_boundary(X, M, 'Support Vector Machine dengan Polynomial Kernel of Degree 3')
```

### 12. Kita coba set parameter degree Polynomial Kernel menjadi lebih tinggi yaitu dengan nilai 5

```python
M = svm.SVC(kernel='poly', degree=5, gamma='auto', C=10000)
M.fit(X, y)
M.score(X, y)
```

```python
plot_data_and_boundary(X, M, 'Support Vector Machine dengan Polynomial Kernel of Degree 5')
```
