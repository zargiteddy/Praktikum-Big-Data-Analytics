### 4. Menentukan ukuran sebaran data

Menghitung nilai deviasi standar sebuah variabel dari data set. Lakukan edit coding pada praktek 1 (memanggio dataset) dengan menggunakan “dataframe”. Hal ini untuk memudahkan pencarian nilai dari ukuran statistic deskriptif.

```python
import pandas as pd
data = pd.read_csv("brain_size.csv", sep=';', na_values=".")
```

```python
df = data
df
```

Setelah itu kita buat coding untuk rumus deviasi standart

```python
def stdev(nums):
    diffs = 0
    avg = sum(nums)/len(nums)
    for n in nums:
        diffs += (n - avg)**(2)
    return (diffs/(len(nums)-1))**(0.5)
```

Misal akan dicari nilai deviasi standart dari variabel VIQ dengan cara membuat inisiasi variabel tersebut dan setelah itu kita berikan fungsi deviasi standar yang sdh didefinisikan.

```python
viq = df['VIQ']
viq
```

Apabila ada data dari sebuah variabel yang Null (dalam dataset ditulis NaN) dan kita berikan fungsi stdev maka nilai deviasi standar tidak muncul angka tetapi muncul NaN

```python
data_height = df['Height']
data_height
print(stdev(data_height))
```

Apabila menjumpai hal ini maka data NaN dapat diganti/diestimasi dengan nilai “mean” agar bisa dihitung nilai devstandartnya. Apabila nilai “NaN” telah diganti oleh mean maka hasil nilai deviasi standar dari Height sebagai berikut :

```python
df['Height'] = df['Height'].fillna(df['Height'].mean())
df
```

```python
data_height = df['Height']
data_height
print(stdev(data_height))
```
