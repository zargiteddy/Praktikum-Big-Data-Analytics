### 9. Untuk melakukan Load data Job, buatlah file baru bernama etl-5.py, ketikkan script berikut:

```python
import sqlalchemy as sql
import pandas as pd

conn_string = "mysql+pymysql://root:@localhost/bigdata6"

sql_engine = sql.create_engine(conn_string)

#extract : baca isi pegawai_cab01.csv
ds = pd.read_csv("pegawai_cab01.csv")

#transform : ambil kolom job_id dan job_title
#hilangkan data kembar untuk kedua field tersebut
df = ds[['job_id','job_title']].drop_duplicates()
#load data ke MySQL
df.to_sql("job", con=sql_engine, if_exists='replace', index=False)
```

Mirip seperti _script_ di nomer 7 tadi, hanya saja di sini kita melakukan dua hal baru. Yang pertama yaitu mengambil kolom `job_id` dan `job_title` kemudian kita hilangkan data kembar. Yang kedua adalah kita load datanya ke dalam tabel bernama `job` dengan baris kode `df.to_sql("job", con=sql_engine, if_exists='replace', index=False)`.

Jalankan script tersebut kemudian cek melalui phpMyAdmin apakah proses Load tabel job sudah berhasil
