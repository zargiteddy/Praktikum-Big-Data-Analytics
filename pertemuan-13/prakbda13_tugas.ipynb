## 3. Cari data set bebas dari sumber data kaggle.com dan lakukan visualisasi dari dataset tersebut
(Saya menggunakan dataset Penguin Palmer Archipelago bernama penguins_size.csv yang saya dapatkan dari kaggle. Berikut link dataset tersebut:
https://www.kaggle.com/datasets/parulpandey/palmer-archipelago-antarctica-penguin-data )

Pertama, lakukan praprocessing pada dataset terlebih dahulu untuk mengisi NaN values

Import libraries yang diperlukan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Tampilkan dataset tersebut

```python
penguin = pd.read_csv("penguins_size.csv")
print(penguin.head(10))
```

Lihat NaN values di setiap variabel

```python
penguin.isnull().sum()
```

Note: yang kita perlukan hanyalah variabel culmen_length_mm dan culmen_depth_mm, sehingga kita hanya akan mengisi NaN values pada dua variabel tersebut.

Isi NaN values culmen_length_mm dengan rata-rata:

```python
penguin['culmen_length_mm'] = penguin['culmen_length_mm'].fillna(penguin['culmen_length_mm'].mean())
penguin.head(10)
```

Isi NaN values culmen_depth_mm dengan rata-rata:

```python
penguin['culmen_depth_mm'] = penguin['culmen_depth_mm'].fillna(penguin['culmen_depth_mm'].mean())
penguin.head(10)
```

Setelah selesai proses data praprocessing, kita lakukan visualisasi dengan seaborn.

Visualisasi data:

**a) Scatter Plot**

```python
ax = sns.scatterplot(x='culmen_length_mm', y='culmen_depth_mm', data=penguin)
```

Dengan mengambil dua variabel yaitu `culmen_length_mm (x)` dan `culmen_depth_mm (y)`, kita tampilkan sebaran data dengan fungsi seaborn yaitu scatterplot. Scatter plot menggunakan dot (titik) untuk merepresentasikan nilai dari dua variabel numerik yang diplot pada sisi horizontal (x) dan vertikal (y).

**b) Grafik Garis (Line Plot)**

```python
ay = sns.lineplot(x='culmen_length_mm', y='culmen_depth_mm', data=penguin)
```

Dengan menggunakan dua variabel yang sama dengan tadi, yaitu `culmen_length_mm` dan `culmen_depth_mm`, kita tampilkan grafik datanya menggunakan lineplot. `Lineplot()` adalah salah satu fungsi visualisasi data seaborn yang berguna untuk menampilkan grafik informasi berupa rangkaian titik data yang terhubung dengan segmen garis lurus. Pada kode di atas, kita membuat line plot dengan menggunakan metode sns.lineplot dan memberikan argumen berupa data yang digunakan sebagai sumbu x dan y yaitu `culmen_length_mm` untuk sumbu `x` dan `culmen_depth_mm` untuk sumbu `y`, sehingga kita dapat menuliskan baris kode `ay = sns.lineplot(x='culmen_length_mm', y='culmen_depth_mm', data=penguin)`.

**c) Distribusi Normal**

```python
penguin = np.random.normal(size=100)
sns.distplot(penguin);
```

Untuk menghitung distribusi normal dataset penguins_size, kita tuliskan baris kode `penguin = np.random.normal(size=100)`. Bisa kita perhatikan jika ada parameter size yang merupakan parameter opsional yang berguna untuk menentukan bentuk array yang dihasilkan. Lalu, untuk menampilkan grafik kita menggunakan distplot yaitu fungsi seaborn untuk menampilkan grafik histogram dengan garis yang bisa diaplikasikan dengan menuliskan kode `sns.distplot(penguin);`.
