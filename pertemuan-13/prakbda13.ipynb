Pertama kita deklarasi variabel df untuk mewakili DataFrame. Kemudian di dalamnya kita gunakan fungsi pandas untuk membaca dataset yaitu dengan `pd.read_csv`. Lalu tampilkan 5 data teratas dari dataset `brain_size.csv` dengan menuliskan kode `df.head()`.

### 4. Kita akan tampilkan 2 variabel VIQ dan MRI_Count dari dataset df dengan module scatterplot dari package seaborn

```python
ax = sns.scatterplot(x='VIQ', y='MRI_Count', data=df)
```

Dengan mengambil dua variabel yaitu VIQ (x) dan MRI_Count (y), kita tampilkan sebaran data dengan fungsi seaborn yaitu scatterplot. Scatter plot menggunakan dot (titik) untuk merepresentasikan nilai dari dua variabel numerik yang diplot pada sisi horizontal (x) dan vertikal (y).

### 5. Membuat grafik garis dari seaborn

```python
ay = sns.lineplot(x='VIQ', y='MRI_Count', data=df)
```

Dengan menggunakan dua variabel yang sama dengan tadi, yaitu VIQ dan MRI_Count, kita tampilkan grafik datanya menggunakan lineplot. `Lineplot()` merupakan salahsatu fungsi visualisasi data seaborn yang merupakan jenis plot dasar yang berguna untuk menampilkan grafik informasi berupa rangkaian titik data yang terhubung dengan segmen garis lurus. Pada kode di atas, kita membuat line plot dengan menggunakan metode sns.lineplot dan memberikan argumen berupa data yang digunakan sebagai sumbu x dan y yaitu VIQ untuk sumbu x dan MRI_Count untuk sumbu y, sehingga kodenya dapat ditulis `ay = sns.lineplot(x='VIQ', y='MRI_Count', data=df)`.

### 6. Ketikkan perintah dibawah ini dan grafik apa yang akan muncul

```python
ay = sns.pairplot(data=df)
```

Fungsi pairplot membuat grid sumbu sedemikian rupa sehingga setiap variabel dalam data akan dibagi di sumbu y di satu baris dan di sumbu x di satu kolom. Pair plot digunakan untuk memahami kumpulan fitur terbaik untuk menjelaskan hubungan antara dua variabel atau untuk membentuk cluster yang paling terpisah. Pair plot juga membantu kita untuk membentuk beberapa model klasifikasi sederhana dengan menggambar beberapa garis sederhana atau membuat pemisahan linier dalam dataset.

### 7. Relationship Plot

```python
az = sns.relplot(x='MRI_Count', y='Weight', data=df)
```

Pada kode di atas kita menggunakan relplot yang memiliki fungsi untuk menyediakan akses ke beberapa fungsi sumbu yang berbeda yang menunjukkan hubungan antara dua variabel dengan pemetaan semantik dari subset. Pada kode di atas, kita menunjukkan relationship di antara dua kolom yaitu MRI_Count dan Weight.
